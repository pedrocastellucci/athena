# Numpy

```python
import numpy as np
b = np.array([(1,2), (2, 1)])

print(b)
print()
print(2*b)
print()
print(b*b)
print()
print(b**2)
print()
print(1/b)  # Warning!
```

```python
a = np.array([1, 2, 3])
b = np.array([2, 3, 4])
a * b, a.dot(b)
```

```python
A = np.matrix([
    [1, 2], 
    [2, 1]
])

w, v = np.linalg.eig(A)
print(w)  # Eigenvalues
print(v)  # Eingenvector
A
```
